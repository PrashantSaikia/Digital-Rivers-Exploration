```python
def gen_river(r, n):
    # r is the starting number for the sequence
    # n is the number of terms to be generated
    lis = [r]
    for i in range(n):
        lis.append(lis[-1] + sum(map(int, str(lis[-1]))))
    return lis
```

```python
river_9 = gen_river(9,1000000)
river_3 = gen_river(3,1000000)
river_1 = gen_river(1,1000000)
```

```python
river_1_inc = []
river_3_inc = []
river_9_inc = []

for i in range(8000):
    river_1_inc.append(river_1[i+1]-river_1[i])
    river_3_inc.append(river_3[i+1]-river_3[i])
    river_9_inc.append(river_9[i+1]-river_9[i])
```

```python
def gen_seq(n):
    '''
    Generates the sequence a_n = a_(n-1) + 1 (which can also be expressed as a_n = n+1)
    - n is the number of terms to generate in the sequence
    '''
    lis = [1] # starting number of the sequence
    for i in range(1,n):
        lis.append(lis[-1] + 2*(i+1) - 1)
    return lis
```

```python
import time
t = time.time()
foo = gen_seq(10000000)
print('Time taken: {}s'.format(time.time()-t))
```

```python
def gen_seq(n):
    a = 0
    for i in range(1, n+1):
        a += 2*i - 1
        yield a
```

```python
import time
t = time.time()
```
